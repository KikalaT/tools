```python
import requests
import pandas as pd
from io import StringIO
from geopy.geocoders import Nominatim
print('Setup complete')
```

```python
data = requests.get('https://github.com/opencovid19-fr/data/raw/master/dist/chiffres-cles.csv')
data = StringIO(data.text)
df = pd.read_csv(data, sep=",", low_memory=False)
print('Data imported.')
```

```python
def get_lat_lon(area):
    locator = Nominatim(user_agent="myGeocoder")
    location = locator.geocode(area+", France")
    coordinates = [location.latitude,location.longitude]
    print(area,coordinates)
    return coordinates
```

```python
d = {}
```

```python
for area in df['maille_nom'].unique():
    coordinates = get_lat_lon(area)
    print(area,coordinates)
    d[area] = coordinates
```
